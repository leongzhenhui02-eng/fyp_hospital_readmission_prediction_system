### DATA VISUALISATION

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# For better plot aesthetics
sns.set(style='whitegrid')

#Loading Dataset
dd = pd.read_csv('diabetic_data_processed.csv')
```

1. Count for Readmission Outcomes

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='readmitted', data=dd, palette='Set2')
plt.title('Distribution of Readmission Outcome')
plt.xlabel('Readmitted')
plt.ylabel('Number of Patients')
plt.show()
```

2. Age vs Readmission

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='age', hue='readmitted', data=dd, palette='Set1')
plt.title('Readmission by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

3. Gender Distribution

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='gender', data=dd, palette='Set3')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

4. Distribution of Number of Medications

```python
plt.figure(figsize=(8, 5))
sns.histplot(dd['num_medications'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Number of Medications')
plt.xlabel('Number of Medications')
plt.ylabel('Frequency')
plt.show()
```

5. Time in hospital vs readmission

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='readmitted', y='time_in_hospital', data=dd, palette='pastel')
plt.title('Time in Hospital by Readmission Status')
plt.xlabel('Readmitted')
plt.ylabel('Time in Hospital (days)')
plt.show()
```

6. Count of Diabetes Medication Change

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='change', data=dd, palette='muted')
plt.title('Count of Medication Change')
plt.xlabel('Change in Diabetes Medication')
plt.ylabel('Number of Patients')
plt.show()
```
